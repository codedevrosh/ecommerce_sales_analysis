## Data Inspection And Preprocessing

The required libraries are imported and data is loaded using Pandas.

```python
import pandas as pd
from pathlib import Path

customers = pd.read_csv(Path("../data/olist_customers_dataset.csv"))
orders = pd.read_csv(Path("../data/olist_orders_dataset.csv"))
order_items = pd.read_csv(Path("../data/olist_order_items_dataset.csv"))
products = pd.read_csv(Path("../data/olist_products_dataset.csv"))
payments = pd.read_csv(Path("../data/olist_order_payments_dataset.csv"))
reviews = pd.read_csv(Path("../data/olist_order_reviews_dataset.csv"))
sellers = pd.read_csv(Path("../data/olist_sellers_dataset.csv"))
category_translation=pd.read_csv(Path("../data/product_category_name_translation.csv"))
```

## Exploratory Data Analysis

```python
sales = orders.merge(order_items, on="order_id")

sales['month'] = sales['order_purchase_timestamp'].dt.to_period('M')

sales.groupby('month')['price'].sum()
```

```python
sales_products = order_items.merge(products, on="product_id")

sales_products.groupby("product_category_name_english")["price"].sum().sort_values(ascending=False).head(10)
```

Health_Beauty generated more revenue.

```python
sales = orders.merge(customers, on="customer_id").merge(order_items, on="order_id")

sales.groupby("customer_state")["price"].sum().sort_values(ascending=False)
```

SP contributes the largest share of marketplace sales.

```python
sales.groupby("order_id")["price"].sum().mean()
```

The average order value on the platform is approximately 140 BRL

```python
orders['delivery_time'] = (
    orders['order_delivered_customer_date'] -
    orders['order_purchase_timestamp']
).dt.days
```

```python
orders['delivery_time'].mean()
```

The average delivery time is around 12 days.

```python
delivery_reviews = orders.merge(reviews, on="order_id")

delivery_reviews.groupby("review_score")["delivery_time"].mean()
```

Orders with longer delivery times tend to receive lower review scores.

## Key Insights

**Customer Distribution:**

* The marketplace has 99,941 customers, with the majority located in São Paulo, followed by Rio de Janeiro.
* This indicates that large urban regions dominate customer demand on the platform.

Business meaning:
Marketing and logistics strategies should focus heavily on these high-demand regions.

**Order Performance:**

* Most orders in the dataset have the status “delivered”, indicating that the platform maintains a high order completion rate.

* Some timestamp fields (delivery or carrier dates) contain missing values because canceled or undelivered orders do not progress through the entire order lifecycle.

Business meaning:
The platform generally maintains reliable order fulfillment.

**Product Availability:**

The most common product categories in the marketplace are:

* Bed, Bath & Table
* Sports & Leisure
* Furniture & Decor
* Beauty & Health
* Household Utilities

These categories suggest that home and lifestyle products dominate the platform inventory.

Business meaning:
The marketplace is heavily oriented toward home improvement and lifestyle products.

**Pricing Insights:**

* The average product price in orders is approximately 120 BRL, indicating that the marketplace contains mostly mid-range priced items rather than luxury products.

Business meaning:
The platform primarily targets middle-income consumers.

**Payment Behavior:**

* The majority of transactions are made using credit cards, while other payment methods such as boleto or debit card are used less frequently.
* Some payments are made in installments, which is common in Brazilian e-commerce.

Business meaning:
Flexible payment options like credit cards and installments play a key role in driving purchases.

**Customer Satisfaction:**

* Review scores show that most customers leave positive ratings, indicating a generally satisfactory shopping experience.

Business meaning:
High ratings suggest that product quality and service meet customer expectations.

**Seller Distribution:**

* Most sellers are located in São Paulo, which aligns with the high concentration of customers in the same region.

Business meaning:
* The marketplace ecosystem is geographically concentrated, which may help with logistics efficiency.


* Overall, the analysis shows that the marketplace is dominated by urban customers, home-related product categories, and credit-card based transactions.
* The platform maintains high order completion rates and generally positive customer satisfaction, indicating a stable and mature e-commerce ecosystem.
